```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
from PIL import Image
import numpy as np
```

```python
f=open('/content/drive/MyDrive/Colab Notebooks/wikipedia.txt', 'r', encoding='UTF8')
text=f.read()
print(text)
```

```python
wc=WordCloud()
wc.generate(text)
plt.imshow(wc)
plt.show()
```

```python
=STOPWORDS.union({'','','',''})
maskHeart=np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/heart_mask.jpg'))
path='/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf'
wc=WordCloud(width=1500, height=1000,
             stopwords=,
             background_color='white',
             mask=maskHeart,
             font_path=path)
wc.generate(text)
plt.imshow(wc)
plt.axis('off')
plt.show()
```
