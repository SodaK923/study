```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS
```

```python
STOPWORDS
```

```python
=STOPWORDS.union({'field','human','search'})
wc_output=WordCloud(width=1500, height=1000, stopwords=, background_color='white')
wc_output.generate(text)
plt.imshow(wc_output)
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import numpy as np
```

```python
=STOPWORDS.union({'field','human','search'})
maskHeart=np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/heart_mask.jpg'))
wc_output=WordCloud(width=1500, height=1000, stopwords=, background_color='white', mask=maskHeart)
wc_output.generate(text)
plt.imshow(wc_output)
plt.axis('off')
plt.show()
```
